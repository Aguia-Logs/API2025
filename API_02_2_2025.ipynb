```python
# Leitura dos arquivos CSV
df1 = pd.read_csv('/content/drive/MyDrive/DadosComexStat/EXP-Filt-06.csv', sep=';', encoding='utf-8', low_memory=False)  # contém CO_PAIS e NO_PAIS
df2 = pd.read_csv('/content/drive/MyDrive/DadosComexStat/Pais_Continentes_Arrumado_02.csv', sep=';', encoding='utf-8')  # contém CO_PAIS, NO_PAIS e Continentes

# Junta os DataFrames pela coluna CO_PAIS
df_merged = df1.drop(columns='NO_PAIS').merge(df2, on='CO_PAIS', how='left')

# Exibe o DataFrame resultante
print(df_merged.head(5))
```

```python
df_merged['NO_URF'] = df_merged['NO_URF'].apply(lambda x: x[9:] if isinstance(x, str) else x)

df_merged.head(5)
```
